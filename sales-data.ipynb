2. What are the total sales attributed to customers over the age of 45?

```python
mergeddf["total_rev"]= mergeddf["quantity"]*mergeddf["price"]
over45 = mergeddf.loc[mergeddf['age'] > 45, 'total_rev'].sum()

print(f"The total sales attributed to customers over the age of 45 is {over45}.")
```

3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)

```python
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df["total_rev"] = df["quantity"] * df["price"]

graph = df.groupby(df['invoice_date'].dt.to_period('M'))['total_rev'].sum()

graph.index = graph.index.astype(str)
plt.plot(graph.index, graph.values)
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Month')
plt.xticks(rotation=90)
plt.show()
```

Based on the graph above, the sales from January 2021 to December 2022 were consistently within the range of 8,000,000 to 10,000,000 total revenue. However, it decreased in January 2023 and experienced the deepest dip in March 2023. From March 2023 to December 2023, the total revenue remained below 2,000,000.

1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth).

```python
age_bins = range(0, 101, 10)
age_labels = [f'{i}-{i+9}' for i in range(0, 100, 10)]
mergeddf['Age Group'] = pd.cut(mergeddf['age'], bins=age_bins, labels=age_labels)

mergeddf
```

```python
pivot_table=pd.pivot_table(mergeddf, values="total_rev",index="category", columns="Age Group",aggfunc="sum")
pivot_table
```
